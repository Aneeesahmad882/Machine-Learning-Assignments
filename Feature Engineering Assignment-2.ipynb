# Answer1

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Sample dataset
data = np.array([[1.0, 2.0],
                 [5.0, 8.0],
                 [2.0, 6.0]])

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler on the data and transform it
data_scaled = scaler.fit_transform(data)

# Display the original and scaled data
print("Original Data:\n", data)
print("\nScaled Data:\n", data_scaled)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the data
X_standardized = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Apply PCA with 2 components (reduce to 2 dimensions)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standardized)

# Display the original and reduced-dimension data
print("Original Data Shape:", X_standardized.shape)
print("Reduced Data Shape:", X_pca.shape)
```

# Answer4

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load the digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Apply PCA for feature extraction (reduce to 2 components for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the reduced-dimension data
plt.figure(figsize=(8, 6))
for i in range(10):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=str(i), alpha=0.8)
plt.title('PCA for Feature Extraction (Digits Dataset)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Sample dataset
data = {
    'price': [10.0, 25.0, 15.0, 30.0],
    'rating': [4.5, 3.0, 5.0, 4.2],
    'delivery_time': [20, 40, 25, 35]
}

df = pd.DataFrame(data)

# Display the original data
print("Original Data:\n", df)

# Extract relevant features for scaling
features_to_scale = ['price', 'rating', 'delivery_time']

# Apply Min-Max scaling
scaler = MinMaxScaler()
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

# Display the scaled data
print("\nScaled Data:\n", df[features_to_scale])
```

# Answer8

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generating a synthetic dataset
np.random.seed(42)
height = np.random.normal(160, 10, 100)
weight = 0.7 * height + np.random.normal(0, 5, 100)

# Creating a feature matrix
X = np.column_stack((height, weight))

# Standardizing the data (optional but often recommended)
mean_vals = np.mean(X, axis=0)
std_dev = np.std(X, axis=0)
X_standardized = (X - mean_vals) / std_dev

# Applying PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Plotting the explained variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```
